```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Data Set
data = pd.read_excel ("/Users/cheryl/Downloads/world_population_data.xlsx")

# print first few rows of the dataset
print(data.head())
```

```python
# set the figure aesthatics
sns.set(style="whitegrid")
```

```python
# Bar Chart: Top 10 countries by Population
plt.figure(figsize=(10,6))
top_10_countries = data.nlargest(10, 'population')
sns.barplot(x='country', y='population',data=top_10_countries,palette='Blues_d')
plt.title('Top Countries by Population')
plt.xlabel('Country')
plt.ylabel('Population(in billions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 1. Select 20 Countries 
top_20_countries = data.nlargest(20, '2022 population')  # Selecting the top 20 by 2022 population

# Line Graph: 2022 Population vs Density, Growth Rate, and World Percentage (for 20 countries)
plt.figure(figsize=(12, 7))

variables = ['density (km²)', 'growth rate', 'world percentage']

# Plot each variable for the selected 20 countries
for variable in variables:
    plt.plot(top_20_countries['country'], top_20_countries[variable], label=variable)

# Set the labels, title, and legend
plt.title('2022 Population vs Other Variables (Density, Growth Rate, World Percentage) - Top 20 Countries')
plt.xlabel('Country')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: Area (km²) vs 2015 Population, colored by Continent, sized by Rank
plt.figure(figsize=(12, 7))

# Create the scatter plot
scatter = sns.scatterplot(
    x='area (km²)', 
    y='2015 population', 
    hue='continent',  
    size='rank',  
    sizes=(200, 500),  
    data=data
)

# Set the title and labels
plt.title('Scatter Plot of Area (km²) vs 2015 Population by Continent and Rank')
plt.xlabel('area (km²)')
plt.ylabel('2015 Population')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  

# Show the plot
plt.tight_layout()
plt.show()
```

### Bar Graph
The chart shows that China and India have the largest population by a wide margin. While other countries like USA, Brazil, Nigeria and Indonesia are significantly smaller as compared to the first 2 countries. This highlights the importance of considering both absolute population numbers and growth rates when studying population dynamics.

### Line Graph 
I used nlargest (20, 2022 population) to select the top 20 countries based on population in 2022. It shows the realtionship between Population density, growth rate and world percentage for the selected countries.Population growth is strongly linked to density.

### Scatter Plot
This gives a visual represantation of how population and land area compare across different continents, while also reflecting on the country rating. This can show if larger countries tend to have bigger populations or higher ranks and if this differs across different continents. There is a noticeable inverse relationship between population size and area covered. Asia despite its large populations continues to grow at significant rates.
