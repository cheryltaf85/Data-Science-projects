# Adult Income Data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Read the Adult Income Dataset from the Given URL

```python
# Load the dataset
data = pd.read_csv('/Users/cheryl/Downloads/adult_income_data.csv')

# Check the current column names and number of columns
print(data.columns)
print(len(data.columns))  # This will tell you how many columns are present
```

### Create a Script That Will Read a Text File Line by Line

```python
# Assign the correct number of column names (based on 14 columns)
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 
                'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 
                'hours_per_week', 'native_country']
```

### Add a Name of Income for the Response Variable

```python
# Preview the dataset
print(data.head())
```

```python
# Function to read a file line by line
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            print(line.strip())
```

### Find the Missing Values

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

### Create a DataFrame with Only Age, Education, and Occupation:

```python
# Subset with age, education, and occupation
subset_df = data[['age', 'education', 'occupation']]

# Preview the new DataFrame
print(subset_df.head())
```

### Plot a Histogram of Age with a Bin Size of 20

```python
# Plot histogram of age
plt.hist(data['age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Create a Function to Strip Whitespace Characters

```python
# Function to strip whitespace
def strip_whitespace(column):
    return column.str.strip()
```

### Apply This Function to All Columns with String Values

```python
# Function to strip whitespace from a string
def strip_whitespace(value):
    if isinstance(value, str):
        return value.strip()
    return value  # If not a string, return as is

# Apply the function to all string columns
for column in data.select_dtypes(include=['object']).columns:
    data[column] = data[column].apply(strip_whitespace)

# Check the DataFrame after applying the function
print(data.head())
```

### Find the Number of People Who Are Aged Between 30 and 50

```python
# Filter for people aged between 30 and 50
age_30_50 = data[(data['age'] >= 30) & (data['age'] <= 50)]
print(f"Number of people aged between 30 and 50: {len(age_30_50)}")
```

### Group by Age and Education to Find How Mean Age is Distributed

```python
# Group by age and education to calculate the mean
mean_age_distribution = data.groupby(['age', 'education'])['age'].mean()
print(mean_age_distribution)
```

### Group by Occupation and Show Summary Statistics of Age

```python
# Group by occupation and show summary statistics of age
age_summary_by_occupation = data.groupby('occupation')['age'].describe()
print(age_summary_by_occupation)

# Find which profession has the oldest workers on average
oldest_workers = age_summary_by_occupation['mean'].idxmax()
print(f"Profession with the oldest workers on average: {oldest_workers}")

# Find which profession has the largest share of workforce above the 75th percentile
largest_above_75th = age_summary_by_occupation['75%'].idxmax()
print(f"Profession with the largest share above the 75th percentile: {largest_above_75th}")
```

### Use Subset and GroupBy to Find the Outliers:

```python
# Define outliers as those beyond the 1.5*IQR rule
Q1 = data['age'].quantile(0.25)
Q3 = data['age'].quantile(0.75)
IQR = Q3 - Q1

outliers = data[(data['age'] < (Q1 - 1.5 * IQR)) | (data['age'] > (Q3 + 1.5 * IQR))]

# Group outliers by occupation
outliers_by_occupation = outliers.groupby('occupation')['age'].count()
print(outliers_by_occupation)
```

### Plot the Outlier Values on a Bar Chart:

```python
# Plot the outliers on a bar chart
outliers_by_occupation.plot(kind='bar')
plt.title('Outliers by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Number of Outliers')
plt.show()
```

### Merge the Two DataFrames Using Common Keys to Drop Duplicate Values

```python
# Assuming we have two DataFrames df1 and df2 with common keys
merged_df = pd.merge(data, subset_df, how='inner', on=['age', 'education', 'occupation'])

# Drop duplicates
merged_df = merged_df.drop_duplicates()

# Preview the merged DataFrame
print(merged_df.head())
```

# Practising Basic Arithmetic using a series

```python
import pandas as pd
```

```python

# Create Series 1
series1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
```

```python
# Create Series 2
series2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
```

```python
# Add Series 1 and Series 2
addition_result = series1 + series2
print("Addition of Series 1 and Series 2:")
print(addition_result)
```

```python
# Subtract Series 1 from Series 2
subtraction_result = series2 - series1
print("\nSubtraction of Series 1 from Series 2:")
print(subtraction_result)
```
